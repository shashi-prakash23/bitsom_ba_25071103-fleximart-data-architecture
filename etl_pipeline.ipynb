```python
import re
import unicodedata

# -----------------------------
# Helper dictionaries & patterns
# -----------------------------

# Common product noun normalization
PRODUCT_SYNONYMS = {
    r"\b(t[-\s]?shirt|tees?)\b": "tshirt",
    r"\b(track\s?pants?)\b": "trackpants",
    r"\b(head[-\s]?phones)\b": "headphones",
    r"\b(ear[-\s]?buds?)\b": "earbuds",
    r"\b(jeans)\b": "jeans",
    r"\b(sneakers|shoes)\b": "sneakers",
    r"\b(lap[-\s]?top)\b": "laptop",
    r"\b(tv|television)\b": "tv",
    r"\b(monitor)\b": "monitor",
    r"\b(rice)\b": "rice",
    r"\b(almonds)\b": "almonds",
    r"\b(honey)\b": "honey",
    r"\b(shirt)\b": "shirt",
    r"\b(ear phones?)\b": "earphones",
}

# Category normalization
CATEGORY_SYNONYMS = {
    r"^electronics$": "Electronics",
    r"^fashion$": "Fashion",
    r"^groceries$": "Groceries",
}

# Units normalization patterns
INCH_PAT = re.compile(r"(\d+)\s*(in|inch|inches)\b", flags=re.I)
KG_PAT = re.compile(r"(\d+(?:\.\d+)?)\s*(kg|kilograms?)\b", flags=re.I)
G_PAT = re.compile(r"(\d+)\s*(g|grams?)\b", flags=re.I)
ML_PAT = re.compile(r"(\d+)\s*(ml|milliliters?)\b", flags=re.I)

# Known nouns to separate brand from product
KNOWN_NOUNS = [
    "galaxy", "nike", "macbook", "jeans", "headphones", "almonds", "laptop",
    "tshirt", "rice", "nord", "sneakers", "monitor", "tv", "iphone", "honey",
    "shirt", "earbuds", "earphones", "trackpants"
]

# -----------------------------
# Normalization utilities
# -----------------------------

def normalize_unicode(text: str) -> str:
    """Normalize unicode and remove non-printable control chars."""
    if pd.isna(text):
        return text
    # NFKC to normalize width variants; strip control characters
    text = unicodedata.normalize("NFKC", str(text))
    text = re.sub(r"[\u200B-\u200D\uFEFF]", "", text)  # zero-width spaces
    return text

def clean_whitespace(text: str) -> str:
    """Trim and collapse multiple spaces."""
    text = text.strip()
    text = re.sub(r"\s+", " ", text)
    return text

def normalize_units(text: str) -> (str, dict):
    """Standardize units to canonical forms and return extracted attributes."""
    attrs = {}
    # inches
    def _inch_repl(m):
        val = int(m.group(1))
        attrs["size_in"] = val
        return f"{val} in"
    text = INCH_PAT.sub(_inch_repl, text)

    # kg
    def _kg_repl(m):
        val = float(m.group(1))
        attrs["weight_kg"] = val
        return f"{val:g} kg"
    text = KG_PAT.sub(_kg_repl, text)

    # g
    def _g_repl(m):
        val = int(m.group(1))
        attrs["weight_g"] = val
        return f"{val} g"
    text = G_PAT.sub(_g_repl, text)

    # ml
    def _ml_repl(m):
        val = int(m.group(1))
        attrs["volume_ml"] = val
        return f"{val} ml"
    text = ML_PAT.sub(_ml_repl, text)

    # common attached forms: 43inch -> 43 in ; 1kg -> 1 kg ; 500g -> 500 g
    text = re.sub(r"\b(\d+)(inch|inches|in)\b", lambda m: f"{m.group(1)} in", text, flags=re.I)
    text = re.sub(r"\b(\d+(?:\.\d+)?)(kg)\b", lambda m: f"{m.group(1)} kg", text, flags=re.I)
    text = re.sub(r"\b(\d+)(g)\b", lambda m: f"{m.group(1)} g", text, flags=re.I)
    text = re.sub(r"\b(\d+)(ml)\b", lambda m: f"{m.group(1)} ml", text, flags=re.I)
    return text, attrs

def normalize_synonyms(text: str) -> str:
    """Apply product synonym rules (works best on lowercased text)."""
    t = text
    for pat, repl in PRODUCT_SYNONYMS.items():
        t = re.sub(pat, repl, t, flags=re.I)
    return t

def canonical_case(text: str) -> str:
    """Lowercase canonical representation."""
    return text.lower()

def display_case(text: str) -> str:
    """Title-case for display while respecting units and acronyms."""
    # Title-case, then fix 'Tv' -> 'TV', 'In' in units
    t = text.title()
    t = re.sub(r"\bTv\b", "TV", t)
    t = re.sub(r"\bIn\b", "in", t)  # keep unit 'in' lowercase
    return t

def extract_brand_product(canonical_name: str) -> (str, str):
    """
    Naive split: brand is the leading token(s) until a known noun appears.
    e.g., 'samsung galaxy s21' -> brand 'samsung', product 'galaxy s21'
    """
    tokens = canonical_name.split()
    if not tokens:
        return None, canonical_name

    # find first known noun index
    idx = None
    for i, tok in enumerate(tokens):
        if tok in KNOWN_NOUNS:
            idx = i
            break

    if idx is None or idx == 0:
        # brand is the first token if it looks like a brand (capitalized in original)  fallback simple rule
        return tokens[0], " ".join(tokens[1:]) if len(tokens) > 1 else None
    else:
        brand = " ".join(tokens[:idx]) if idx > 0 else None
        product = " ".join(tokens[idx:])
        return brand or None, product or None

def rationalize_name(raw_name: str) -> dict:
    """
    Full pipeline: normalize unicode/whitespace, standardize units,
    harmonize synonyms, produce canonical & display names, and extract brand/product.
    """
    if pd.isna(raw_name) or not str(raw_name).strip():
        return {
            "canonical_name": None,
            "display_name": None,
            "brand": None,
            "product_core": None,
            "size_in": None,
            "weight_g": None,
            "weight_kg": None,
            "volume_ml": None,
        }

    t = normalize_unicode(raw_name)
    t = clean_whitespace(t)

    # Keep punctuation like apostrophes; normalize dashes/underscores to space
    t = re.sub(r"[_|]+", " ", t)
    t = re.sub(r"[-]+", "-", t)  # unify dashes

    # Standardize units and collect attributes
    t, attrs = normalize_units(t)

    # Lowercase for canonical ops, then apply synonyms
    t_lower = canonical_case(t)
    t_lower = normalize_synonyms(t_lower)

    # Remove redundant multiple spaces after substitutions
    t_lower = clean_whitespace(t_lower)

    # Canonical & display
    canonical = t_lower
    display = display_case(canonical)

    # Extract brand & product core (best-effort)
    brand, product_core = extract_brand_product(canonical)

    return {
        "canonical_name": canonical,
        "display_name": display,
        "brand": brand,
        "product_core": product_core,
        "size_in": attrs.get("size_in"),
        "weight_g": attrs.get("weight_g"),
        "weight_kg": attrs.get("weight_kg"),
        "volume_ml": attrs.get("volume_ml"),
    }

def normalize_category(raw_cat: str) -> str:
    if pd.isna(raw_cat):
        return None
    t = normalize_unicode(str(raw_cat))
    t = clean_whitespace(t).lower()
    for pat, repl in CATEGORY_SYNONYMS.items():
        if re.match(pat, t, flags=re.I):
            return repl
    # Title-case fallback
    return t.title()


data = Product_df
df = pd.DataFrame(data)

# Apply rationalization
r_cols = df["product_name"].apply(rationalize_name).apply(pd.Series)
df["canonical_name"] = r_cols["canonical_name"]
df["display_name"] = r_cols["display_name"]
df["brand"] = r_cols["brand"]
df["product_core"] = r_cols["product_core"]
df["size_in"] = r_cols["size_in"]
df["weight_g"] = r_cols["weight_g"]
df["weight_kg"] = r_cols["weight_kg"]
df["volume_ml"] = r_cols["volume_ml"]

# Normalize categories
df["category_normalized"] = df["category"].apply(normalize_category)

# Optional: derive a "rationalized_product_name" you can store back
df["rationalized_product_name"] = df.apply(
    lambda r: r["display_name"], axis=1
)

# Preview
print(df[[
    "product_id", "product_name", "rationalized_product_name",
    "canonical_name", "brand", "product_core",
    "size_in", "weight_g", "weight_kg",
    "category", "category_normalized"
]].to_string(index=False))

Product_df_n_1=df
```

```python

# 1) Identify products with missing price (NaN or 0)
missing_mask = Product_df_n_1['price'].isna() | (Product_df_n_1['price'] == 0)
missing_product_ids = Product_df_n_1.loc[missing_mask, 'product_id'].unique()

# 2) Filter relevant sales rows and keep only those with a unit_price
sales_relevant = Sales_df_n[
    Sales_df_n['product_id'].isin(missing_product_ids) & 
    Sales_df_n['unit_price'].notna()
].copy()

# 3) use the latest unit_price per product_id
if 'sale_date' in sales_relevant.columns:
    # Ensure sale_date is datetime
    sales_relevant['sale_date'] = pd.to_datetime(sales_relevant['sale_date'], errors='coerce')
    # Sort by product_id and sale_date to get most recent
    sales_relevant.sort_values(['product_id', 'sale_date'], inplace=True)
    latest_prices = sales_relevant.dropna(subset=['sale_date']).groupby('product_id').tail(1)
    price_map = latest_prices.set_index('product_id')['unit_price']
else:
    # Fallback: use the median unit_price per product_id (robust against outliers)
    price_map = sales_relevant.groupby('product_id')['unit_price'].median()

# 4) Fill only the missing prices in Products_df_n_1
Product_df_n_1.loc[missing_mask, 'price'] = Product_df_n_1.loc[missing_mask, 'product_id'].map(price_map)

# 5) Optional: report any product_ids still missing (no price found in sales)
still_missing = Product_df_n_1.loc[Product_df_n_1['price'].isna() | (Product_df_n_1['price'] == 0), 'product_id'].unique()
if len(still_missing) > 0:
    print("No unit_price found in sales for these product_id(s):", still_missing)

```
